## Converting duration into appropriate format

```python
from datetime import datetime, time
def change(l):
    m=0
    t=""
    s=0
    h=0
    l=l[2:]
    for i in l:
        if i=='D':
            t=""
            continue
        if i=='T':
            t=""
            continue
        if i=='H':
            h=int(t)
            t=""
            continue
        if i=='M':
            m=int(t)
            t=""
            continue
        if i=='S':
            s=int(t)
            t=""
        t+=i
    duration=time(h,m,s)
    return duration

df['duration']=df['duration'].apply(change)
df.head(2)
```

```python
def total_seconds(i):
    t = i.hour * 3600 + i.minute*60 + i.second
    return t
df['duration']=df['duration'].apply(total_seconds)
df.head(2)
```

## **Converting category_id to Category title**

getting title for each id via web scraping


```python
import requests
from bs4 import BeautifulSoup as bs

r=requests.get('https://mixedanalytics.com/blog/list-of-youtube-video-category-ids/')

soup=bs(r.content)
```

```python
table=soup.find('table')
rows=table.findAll('td')
d={}
for i,v in enumerate(rows):
    if i%2==0:
        try:
            d[int(v.getText())]=rows[i+1].getText()
        except:
            continue
print(d)
```

```python
def title(i):
    if i in d:
        return d[i]
df['category_id']=df['category_id'].apply(title)
df.rename(columns= {'category_id':'category'},inplace=True)
df.head(2)
```

## **Changing publish_time to categorical feature**

- 4:00-11:00 --> Morning
- 11:00-16:00 --> Afternoon
- 16:00-21:00 --> Evening
- 21:00-4:00 --> Night



```python
def change1(i):
    if 4<=i.hour<11:
        return 'Morning'
    elif 11<=i.hour<16:
        return 'Afternoon'
    elif 16<=i.hour<22:
        return 'Evening'
    else:
        return 'Night'

df['publish_time']=df['publish_time'].apply(pd.to_datetime)
df['publish_time']=df['publish_time'].apply(change1)
df.head(2)
```
