```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./datahack_videos.csv")
tags = df["tags"]
tags = np.array(tags)
category = df["category"]
category = np.array(category)
tags[0:5]
```

```python
unique_category = np.unique(category)
freq = {i:{} for i in unique_category}
for i in range(len(tags)):
    tag_array = tags[i][2:-2].split(',')
    for j in tag_array:
        # print(j)
        if j in freq[category[i]].keys():
            freq[category[i]][j] += 1
        else:
            freq[category[i]][j]=1
freq
    
```

```python
best_tags = {}
for i in freq.keys():
    x = sorted(list(freq[i].keys()), key = lambda x: x in freq[i].values())
    best_tags[i] = x[0:5]
best_tags
```
